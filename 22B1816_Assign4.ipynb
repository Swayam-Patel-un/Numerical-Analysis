```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
pi=4*m.atan(1)
tol=1e-6
```

```python
def inte(x,c):
    if c==1:
        return np.exp(-x**2)
    elif c==2:
        return np.log(1+x)
    elif c==3:
        return np.sin(x)
    elif c==4:
        return np.sqrt(x)/(x+4)
    elif c==5:
        return (np.sin(x))**2
    elif c==6:
        return 1/np.sqrt(1+x**2)
```

```python
def g(gc,c,a,b,n):
    if gc==1:
        xi,wi= np.polynomial.legendre.leggauss(n)
        x=(b-a)/2*xi+(a+b)/2
        dx=(b-a)/2
        return dx*np.sum(wi*inte(x,c))
    elif gc==2:
        xi,wi = np.polynomial.laguerre.laggauss(n)
        return np.sum(wi*inte(xi,c))
    elif gc==3:
        xi,wi=np.polynomial.hermite.hermgauss(n)
        return np.sum(wi*inte(xi,c))
```

```python
intn=["Gauss-Legendre quadrature","Gauss-Laguerre quadrature","Gauss-Hermite quadrature"]
def pl(gc,c,a,b):
    n=1
    N=[1]
    Ival=[g(gc,c,a,b,1)]
    while(abs(g(gc,c,a,b,n+1)-g(gc,c,a,b,n))>tol):
        n+=1
        N.append(n)
        Ival.append(g(gc,c,a,b,n))
    N.append(n+1)
    Ival.append(g(gc,c,a,b,N[-1]))
    print("Convergence of the function using",intn[gc-1],"is ",Ival[-1]," at N value being",n+1)
    return N,Ival
```

```python
x,y=pl(1,1,-1,1)
plt.plot(x,y,'o',linestyle='-')
plt.xlabel("Number of Quadrature points")
plt.ylabel("Integral value")
```

```python
x,y=pl(1,2,0,pi/2)
plt.plot(x,y,'o',linestyle='-')
plt.xlabel("Number of Quadrature points")
plt.ylabel("Integral value")
```
