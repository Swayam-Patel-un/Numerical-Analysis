```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(np.pi*x)
def soln(x,t):
    return f(x)*np.exp(-np.pi*t)
```

```python
def u(N,M,tm):
    umat=np.zeros((N+1,M+1))
    ex=np.zeros((N+1,M+1))
    h=1/N
    k=tm/M
    sig=k/pow(h,2)
    for j in range(M+1):
        for i in range(N+1):
            ex[i][j]=soln(i*h,j*k)
            if j==0:
                umat[i][j]=f(i*h)
            elif i==0 or i==N:
                umat[i][j]=0
            else:
                umat[i][j]=sig*(umat[i+1][j-1]+umat[i-1][j-1])+(1-2*sig)*umat[i][j-1]
    return sig,umat,ex,
    
```

```python
def tabular(U,N,M,tm):
    h=1/N
    k=tm/M
    df=pd.DataFrame(U)
    print('Each row represents the bin along the x axis ')
    print('Each column represents bin along t axis ')
    col=[]
    row=[]
    for step in range(0,U.shape[1]):
        col.append(k*step)
    for step in range(0,U.shape[0]):
        row.append(h*step)
    df.columns=col
    df.index=row
    print(df)  
```

```python
print("This is for h=0.1 and k=0.005, with tmax=1s")
sig,mat,ex=u(10,200,1)
print(sig)
print(mat)
print(ex)
```

```python
def main():
    c='y'
    while (c.lower()=='y'):
        tm=float(input("Enter the maximum time:"))
        N=int(input("Enter the number of intervals in distance:"))
        M=int(input("Enter the number of intervals for time:"))
        sig,mat,ex=u(N,M,tm)
        print("The sigmavalue is",sig)
        print("\nThis is for Numerical method value.")
        tabular(mat,N,M,tm)
        print("\nThis is for Exact values value.")
        tabular(ex,N,M,tm)    
        c=input("Do you want to continue with different values (Y for yes and N for no):")
```
